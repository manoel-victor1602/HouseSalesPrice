### Importação das bibliotecas

```python
# Funções para pré-processamento e preenchimento de dados
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
```

```python
# Funções para cálculo da Curtose e Assimetria dos resíduos
from scipy.stats import kurtosis  
from scipy.stats import skew
```

```python
# Carregando os atributos preditores, atributo alvo e os nomes dos atributos que apresentaram importância para o atributo alvo
X_train = np.load('dataset/X_processed.npy')
y_train = np.load('dataset/y.npy')
features = np.load('dataset/usefull_features.npy', allow_pickle=True)
```

```python
# Carregando a base de teste
df_teste = pd.read_csv('dataset/test.csv', index_col='Id')
df_teste = df_teste[features]
```

```python
# Verificando os primeiros elementos da base de teste
df_teste.head()
```

```python
# Verificando o número de exemplos e colunas presentes na base de teste
df_teste.shape
```

```python
# Alterando o tipo de dado do atributo MSSubClass para categórico
df_teste['MSSubClass'] = df_teste['MSSubClass'].astype('object')
```

```python
df_teste.info()
```

```python
isnull = df_teste.isnull().sum()
isnull_index = [index for null, index in zip(isnull, isnull.index) if null > 0]
```

```python
df_teste[isnull_index].info()
```

### Tratamento de dados faltantes da base de dados de teste

```python
imputer = SimpleImputer(strategy='constant', fill_value=0)
df_teste['MasVnrArea'] = imputer.fit_transform(np.array(df_teste['MasVnrArea']).reshape(-1, 1))
```

```python
garage_isnull = df_teste['GarageYrBlt'].isnull()
df_teste['GarageYrBlt'] = [df_teste.loc[index, 'GarageYrBlt']  if not(null) else df_teste.loc[index, 'YearBuilt'] 
                                                               for null, index in zip(garage_isnull, garage_isnull.index)]
```

```python
isnull = df_teste.isnull().sum()
isnull_index = [index for null, index in zip(isnull, isnull.index) if null > 0]
```

```python
df_teste[isnull_index].info()
```

Na base de teste há atributos com dados faltantes diferentes dos presentes na base de treinamento, como não foi testado qualquer método de preenchimento de dados nessas colunas e para manter a consistência do processo foi decidido por excluir os exemplos com dados ausentes nestes atributos. Sendo eles <i>SaleType</i>, <i>MSZoning</i>, <i>Exterior1st</i> <i>KitchenQual</i>, <i>BsmtFullBath</i>, <i>GarageCars</i> e <i>TotalBsmtSF</i>.

```python
for item in ['SaleType', 'MSZoning','Exterior1st','KitchenQual','BsmtFullBath','GarageCars','TotalBsmtSF']:
    isnull_index.remove(item)
```

```python
isnull_index
```

```python
for col in isnull_index:
    imputer = SimpleImputer(strategy='most_frequent')
    df_teste[col] = imputer.fit_transform(np.array(df_teste[col]).reshape(-1, 1))
```

```python
df_teste = df_teste.dropna()
```

```python
df_teste.info()
```

### Transformação dos atributos categóricos

```python
# Separando a base de dados em atributos preditores (X) e atributo alvo (y)
X_test = df_teste.copy()
```

```python
is_category = df_teste.dtypes == 'object'
cat_index = X_test.columns[is_category]
```

```python
X_test[cat_index].head()
```

```python
# Verificando quantas colunas o conjunto de atributos preditores X deve ter ao final da transformação
is_category_X = X_test.dtypes == 'object'
cat_index = X_test.columns[is_category_X]

print('O conjunto de atributos preditores deve possuir {} colunas ao final da transformação.'.format(X_test[X_test.columns[is_category_X]].nunique().sum() + len(X_test.columns) - np.sum(is_category_X)))
```

```python
# Transformando os atributos preditores
for cat, index in zip(is_category, is_category.index):
    if(cat):
        dum_df = pd.get_dummies(df_teste.loc[:, index], prefix=index, drop_first=True)  # Evitando o Dummy Variable Trap
        X_test = X_test.join(dum_df)                  
```

```python
X_test.head()
```

```python
# Removendo as colunas originais dos atributos categóricos
X_test = X_test.drop(cat_index, axis=1)
```

Foi notado que os atributos preditores após o pré-processamento na base de teste apresentaram diferença em relação a base de terino, isto se dá porque nem todas as categorias presentes dos atributos categóricos na base de treino foram contemplados na base de teste, porém para que se possa realizar os testes com o modelo inteligente é necessário que a base de teste tenha o mesmo número de colunas que a base de treino, portanto foi necessário incluir os atributos categóricos da base de treino não contemplados preenchidos com 0 (não ocorrência do evento) na base de teste.

```python
X_test.head()
```

```python
X_train_df = pd.read_csv('dataset/train_processed_df.csv', index_col='Id')
X_train_df.head()
```

```python
X_test = X_test.reindex(columns = X_train_df.columns, fill_value=0)
```

```python
X_test.head()
```

Agora a base de teste já tem o mesmo formato da base de treino e é possível realizar o treinamento e análise do modelo.

### Treinando o modelo com os dados de treino

```python
# Definindo o modelo Gradient Boosting com os melhores parâmetros encontrados
best_gb = GradientBoostingRegressor(learning_rate= 0.05, 
                                    max_features= 'sqrt',
                                    n_estimators= 1500, 
                                    criterion= 'mae',
                                    loss='huber')
```

```python
# Realizando o treinamento do modelo em toda a base de treino
best_gb.fit(X_train, y_train)
```

### Usando o modelo treinado para previsão dos dados de teste

```python
# Gerando previsões do modelo para toda a base de teste
y_pred = best_gb.predict(X_test)
```

```python
# Visualização da distribuição das previsões do modelo
plt.figure(figsize=(20,10))
sns.distplot(y_pred)
plt.show()
```

```python
qqplot(y_pred, line = 'r')
```

```python
skew(y_pred)
```

```python
kurtosis(y_pred)
```

```python
np.save('dataset/gb_previsoes.npy', y_pred)
```

A distribuição dos valores de venda de casa gerados pelo modelo Gradient Boosting estão ainda apresentando certa normalidade porém com calda mais longa para a direita referente a valores de residências mais altos, a assimetria está maior +1 o que indica que pode estar enviesado e a Curtose Platocúrtica mostra a robustez do modelo contra valores discrepantes.
