### Remoção de Variáveis inicial

```python
# Verificando o número de colunas com dados faltantes

n_columns_missing = 0
for i in df.isnull().sum():
    if(i != 0):
        n_columns_missing += 1
print('Há {} colunas com dados faltantes na base de dados'.format(n_columns_missing))
```

```python
df.isnull().sum()
```

```python
# Verificando quais colunas possuem menos de 10% da amostra total com dados preenchidos

columns_missing = []
for i, n in zip(df.isnull().sum().index, df.isnull().sum()):
    if(n > (len(df) * 0.1)):
        columns_missing.append(i)

print('Há {} colunas que possuem mais de 10% da amostra total com dados faltantes, sendo elas: \n'.format(len(columns_missing)))
for column in columns_missing:
    print('{} com {} elementos faltantes.'.format(column, (df[column].isnull().sum())))
```

```python
df_copy = df.copy()                                # Criando uma cópia do DataFrame
df_copy = df_copy.drop(columns_missing, axis=1)    # Retirando colunas que possuem mais de 10% da amostra total com dados faltantes
```

```python
df_copy.head()                                     # Verificando os 5 primeiros elementos da base de dados
```

```python
df_copy.info()
```

As colunas com muitos dados faltantes foram removidas, mas ainda há alguns atributos que possuem dados faltantes, essas serão tratadas de outras formas em análises futuras. 

```python
# Salvando uma nova base de dados sem as colunas com muitos dados faltantes.
df_copy.to_csv('dataset/train_with_nan.csv')
```

### Treinamento do Modelo Baseline

```python
df_cp_train = df_copy.dropna() # Retirando os exemplos com dados faltantes. 
```

```python
df_cp_train.info()
```

Ao retirar os exemplos e colunas com dados faltantes a base de dados se resumiu a 1338 exemplos e 74 colunas, no entanto a limpeza ainda não foi finalizada e serão feitas mais análises. Mesmo assim, já será criado um modelo de Regressão Linear para gerar um baseline para aferir a qualidade da base de dados.

```python
is_category = df_cp_train.dtypes == 'object'

for cat, index in zip(is_category, is_category.index):
    if(cat):
        df_cp_train.loc[:, index] = LabelEncoder().fit_transform(df_cp_train.loc[:, index])
```

Como o Scikit-learn não aceita que dados categóricos sejam passados para os modelos é necessário transformar essas categorias em números e assim foi aplicado o LabelEncoder, as categorias agora estão representadas por número que vão de 0 à Quantidade de categorias - 1, por exemplo o atributo categórico MSSubClass teve seus valores alterados para números de 0 à 15 (sendo que há 16 categorias presentes).

```python
df_cp_train.info()
```

Após a codificação dos atributos categóricos temos a base de dados sendo representada apenas por números, assim podemos fazer a divisão da base de dados para treino e teste do modelo baseline.

```python
df_cp_train.to_csv('dataset/train_clean_v1.csv') # Salvando a primeira da base de dados própria para modelos inteligentes
```

```python
X = df_cp_train.drop('SalePrice', axis=1)   # O conjunto de atributos preditores inclui todos exceto o atributo alvo (SalePrice)
y = df_cp_train.loc[:, 'SalePrice'].values  # Guardando o atributo alvo em um vetor a parte
```

```python
kf = KFold(n_splits=10)

kf.get_n_splits(X)

mae = []
rmse = []
r2 = []

for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index, :], X.iloc[test_index, :]
    y_train, y_test = y[train_index], y[test_index]
    
    reg = LinearRegression()
    reg.fit(X_train, y_train)
    y_pred = reg.predict(X_test)
    
    print('MAE:  ', mean_absolute_error(y_test, y_pred))
    mae.append(mean_absolute_error(y_test, y_pred))
    
    print('RMSE: ', sqrt(mean_squared_error(y_test, y_pred)))
    rmse.append(sqrt(mean_squared_error(y_test, y_pred)))
    
    print('R2:   ', r2_score(y_test, y_pred))
    r2.append(r2_score(y_test, y_pred))
```

```python
print('A média da métrica MAE foi de {}'.format(round(np.mean(mae), 2)))
print('A média da métrica RMSE foi de {}'.format(round(np.mean(rmse), 2)))
print('A média da métrica R2 foi de {}'.format(round(np.mean(r2), 4)))
```

O modelo baseline obteve bons resultados, mesmo sem ter feito uma limpeza e análise tão aprofundadas na base de dados original, nos próximos notebook serão feitas mais alterações a fim de melhorar as métricas do modelo baseline para quando formos utilizar modelos mais complexos termos uma base de dados de melhor qualidade.
