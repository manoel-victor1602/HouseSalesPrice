### Importação das bibliotecas

```python
import numpy as np # Computação numérica
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

<b>Qqplot</b> - é um gráfico de probabilidades, que é um método gráfico para comparar duas distribuições de probabilidade, traçando seus quantis uns contra os outros.

```python
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.graphics.gofplots import qqplot # 

%matplotlib inline
```

```python
# Funções para cálculo da Curtose e Assimetria dos resíduos
from scipy.stats import kurtosis  
from scipy.stats import skew
```

```python
import warnings
warnings.simplefilter("ignore")
```

```python
# Carregando a base de treinamento processada
X = np.load('dataset/X_processed.npy')
y = np.load('dataset/y.npy')
```

```python
# Definindo o modelo Gradient Boosting com os melhores parâmetros encontrados
best_gb = GradientBoostingRegressor(learning_rate= 0.05, 
                                    max_features= 'sqrt',
                                    n_estimators= 1000, 
                                    criterion= 'mae',
                                    loss='huber')
```

### Análise dos resíduos dos modelos

#### Gradient Boosting

```python
# Gerando previsões para análise posterior
kf = KFold(n_splits=10, shuffle=True)

kf.get_n_splits(X)

gb_y_values = []
gb_y_predicted = []

for train_index, test_index in kf.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    best_gb.fit(X_train, y_train)
    y_pred = best_gb.predict(X_test)
    
    gb_y_values.extend(y_test)
    gb_y_predicted.extend(y_pred)
```

```python
#Cálculo dos resíduos do Gradient Boosting
gb_residuals = np.array(gb_y_values) - np.array(gb_y_predicted)
```

```python
gb_residuals
```

```python
# Visualização da distribuição dos resíduos
plt.figure(figsize=(20,10))
sns.distplot(gb_residuals)
plt.show()
```

A distribuição dos resíduos do Gradient Boosting apresenta o comportamento de uma distribuição normal em torno de 0.

```python
qqplot(gb_residuals, line = 'r')
```

O gráfico Qqplot apresentou a linha de regressão próxima da relação da quantis teóricos e presentes na amostra.

A partir da análise dos gráficos de distribuição dos resíduos e do Qqplot pode-se confirmar que os resíduos estão de acordo com as suposições de homoscedasticidade e normalidade necessários para regressão.

```python
skew(gb_residuals)
```

```python
kurtosis(gb_residuals)
```

A assimetria dos resíduos do Gradient Boosting estão entre -0.5 e +0.5 o que indica que os resíduos são simétricos e a Curtose apresentada é Leptocúrtica, como já demonstrado pelo gráfico de distribuição há uma calda longa, picos altos e possui variação baixa.

#### Regressor Linear

```python
kf = KFold(n_splits=10, shuffle=True)

kf.get_n_splits(X)

lr_y_values = []
lr_y_predicted = []

for train_index, test_index in kf.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    reg_lr = LinearRegression()
    reg_lr.fit(X_train, y_train)
    y_pred = reg_lr.predict(X_test)
    
    lr_y_values.extend(y_test)
    lr_y_predicted.extend(y_pred)
```

```python
lr_residuals = np.array(lr_y_values) - np.array(lr_y_predicted)
```

```python
lr_residuals
```

```python
plt.figure(figsize=(20,10))
sns.distplot(lr_residuals)
plt.show()
```

```python
qqplot(lr_residuals, line = 'r')
```

```python
skew(lr_residuals)
```

```python
kurtosis(lr_residuals)
```

Os resíduos do modelo de Regressão Linear usado como baseline possuem uma assimetria menor -1 o que indica que está enviesado e a Curtose também é Leptocúrtica, porém bem mais variada.

### Conclusão

Os resultados apresentados pelo modelo Gradient Boosting foram bem superiores aos do modelo baseline, o superando em todas as métricas testadas e é o que será usado para geração das previsões da base de teste.
